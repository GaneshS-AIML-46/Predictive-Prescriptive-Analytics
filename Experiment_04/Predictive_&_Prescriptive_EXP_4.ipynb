```python
# The 'tips' dataset is small enough not to require chunking
# The original filtering logic (trip_distance > 0, fare_amount > 0) will be adapted.

df = sns.load_dataset("tips") # Reload for consistency if this cell is run independently

# Filter out rows where total_bill or tip are non-positive, as they wouldn't make sense
df = df[(df["total_bill"] > 0) & (df["tip"] > 0)]

print("Final Shape:", df.shape)
```

```python
plt.figure()
plt.hist(df["total_bill"], bins=100)
plt.title("Histogram of Total Bill")
plt.xlabel("Total Bill")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure()
plt.hist(df["tip"], bins=100)
plt.title("Histogram of Tip Amount")
plt.xlabel("Tip Amount")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure()
plt.scatter(df["total_bill"], df["tip"], alpha=0.3)
plt.xlabel("Total Bill")
plt.ylabel("Tip Amount")
plt.title("Total Bill vs Tip Amount")
plt.show()
```

```python
print(df.describe())
print(df.groupby("size")["tip"].mean())
print(df.groupby("day")[["total_bill", "tip"]].mean())
```
