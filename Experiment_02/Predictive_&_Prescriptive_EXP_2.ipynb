```python
import pandas as pd
import matplotlib.pyplot as plt
import math
```

```python
df = pd.read_csv("/content/drive/MyDrive/Datasets/Mall_Customers.csv")
df.head()
```

```python
annual_income = df["Annual Income (k$)"].values
spending_score = df["Spending Score (1-100)"].values
n = len(annual_income)
```

```python
total_income = 0
for score in annual_income:
    total_income += score
mean_income = total_income / n
print(mean_income)
```

```python
sorted_income = sorted(annual_income)
if n % 2 == 1:
    median_income = sorted_income[n // 2]
else:
    median_income = (sorted_income[n // 2 - 1] + sorted_income[n // 2]) / 2
print(median_income)
```

```python
var_income_sum = 0
for x in annual_income:
    var_income_sum += (x - mean_income) ** 2

variance_income = var_income_sum / n
print(variance_income)
```

```python
std_income = math.sqrt(variance_income)
print(std_income)
```

```python
plt.hist(annual_income, bins=30)
plt.title("Distribution of Annual Income (k$)")
plt.xlabel("Annual Income (k$)")
plt.ylabel("Frequency")
plt.show()
```

```python
sum_spending = 0
for r in spending_score:
    sum_spending += r

mean_spending = sum_spending / n
```

```python
num = 0
den_x = 0
den_y = 0

for i in range(n):
    num += (annual_income[i] - mean_income) * (spending_score[i] - mean_spending)
    den_x += (annual_income[i] - mean_income) ** 2
    den_y += (spending_score[i] - mean_spending) ** 2

correlation = num / math.sqrt(den_x * den_y)
print(correlation)
```

```python
plt.scatter(annual_income, spending_score)
plt.xlabel("Annual Income (k$)")
plt.ylabel("Spending Score (1-100)")
plt.title("Correlation Between Annual Income and Spending Score")
plt.show()
```
